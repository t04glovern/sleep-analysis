```python
import mne
import logging
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (30,5)

logger = logging.getLogger("mne")
logger.setLevel(logging.ERROR)
```

```python
def process_resmed(file_path: str) -> pd.DataFrame:
    """process resmed files
    Args:
        file_path (str): path of the edf files
    Returns:
        pd.DataFrame: returns a DataFrame with breathing data
    """
    edf_file = mne.io.read_raw_edf(file_path, exclude=["Crc16"])

    return edf_file.to_data_frame()
```

```python
edf_file = process_resmed("data/20211030/20211031_030301_PLD.edf")
edf_file.dtypes
edf_file.info
```
